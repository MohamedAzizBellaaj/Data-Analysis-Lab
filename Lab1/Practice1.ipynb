```python
import matplotlib.pyplot as plt
import numpy as np

# p=1/2
x = np.random.binomial(n=1, p=1/2, size=1000)
x_bar = [1/(1+k)*np.sum(x[:k+1]) for k in np.arange(1000)]
plt.plot(np.arange(1, 1001), x_bar, label="X_bar (p=1/2)")
# p=1/3
y = np.random.binomial(n=1, p=1/3, size=1000)
y_bar = [1/(1+k)*np.sum(y[:k+1]) for k in np.arange(1000)]
plt.plot(np.arange(1, 1001), y_bar, label="Y_bar (p=1/3)")
plt.legend()
plt.show()

#! Both x_bar and y_bar converge toward the p parameter of the binomial distribution of x and y respectively.
```
