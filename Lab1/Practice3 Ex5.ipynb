```python
from scipy.stats import shapiro, norm
import matplotlib.pyplot as plt
import numpy as np


sample = np.array([19.41, 16.69, 17.71, 18.15, 21.07, 20.54, 19.52, 18.49,
                   18.69, 18.88, 19.59, 15.10, 20.21, 20.14, 20.14, 21.92,
                   22.64, 21.20, 17.16, 19.75])
plt.hist(sample, bins=5)
plt.xlabel("Charge of cut-off (kg)")
plt.ylabel("Frequency")
plt.show()

#! By the bell-shaped curve of the histogram, this statistical variable likely seems to be normally distributed. But we can verify it further using the Shapiro-Wilk test.
stat, p = shapiro(sample)
if p > 0.05:
    print("The sample appears to be normally distributed.")
else:
    print("The sample does not appear to be normally distributed.")

# Calculate the 95% confidence interval
n = len(sample)
mean = np.mean(sample)
std = np.std(sample, ddof=1)
confidence = 0.95
risk = 1 - confidence
z = norm.ppf(1 - risk / 2)
margin_of_error = z * std / np.sqrt(n)
print("95% Confidence interval:", mean - margin_of_error, mean + margin_of_error)
```
