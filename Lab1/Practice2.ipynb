```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.uniform(size=(10000, 10000))

# Normalization of the first column x0
x0 = x[:, 0]
s0 = (x0-np.mean(x))/np.std(x)
plt.hist(s0)
plt.xlabel("X0")
plt.ylabel("Frequency")
plt.title("Normalized first column visualization")
plt.show()

# Visualization of the central limit theorem
columns_average = np.mean(x, axis=0)
plt.hist(columns_average, bins=20)
plt.xlabel("Sample means")
plt.ylabel("Frequency")
plt.title("Visualization of the central limit theorem")
plt.show()

#! We see that the column average of x has the shape of a normal distribution.
```
